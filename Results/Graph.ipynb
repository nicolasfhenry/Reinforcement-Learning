```python
def plot_reward(results,name,Batch):
    
    data = []
    data = [[i[0] for i in result['reward'] ] for result in results]
    
    plt.figure(figsize = (8,8))
    for d,n in zip(data,name):
        plt.plot(d,label=n)
    plt.title('Comparison of Methods'+'_' + str(Batch),fontsize=20)
    plt.ylabel('Reward',fontsize=20)
    plt.xlabel('Episode',fontsize=20)
    plt.legend()
    plt.savefig('ComparisonOfMethods.png')
    plt.show()
    
    
        
```

```python
Batch = 1
result1 = load_obj('/Acrobots/'+ 'Batch' + str(Batch) + '/' + 'mainGS_mirror_noVBN.py')
result2 = load_obj('/Acrobots/'+ 'Batch' + str(Batch) + '/' +'mainGS_mirror_VBN.py')
result3 = load_obj('/Acrobots/'+ 'Batch' + str(Batch) + '/' + 'mainGS_noMirror_VBN.py')
result4 = load_obj('/Acrobots/'+ 'Batch' + str(Batch) + '/' +'mainGS_noVBN.py')
result5 = load_obj('/Acrobots/'+ 'Batch' + str(Batch) + '/' +'mainGS_noVBN_noFitness.py')


plot_reward([result1,result2,result3,result4,result5],['GS_mirror_noVBN','GS_mirror_VBN','GS_noMirror_VBN','GS_noVBN','GS_noVBN_noFitness'],Batch)
```
